```python
train_data, test_data = houses.random_split(0.8,seed=0)
```

```python
train_data
```

```python
def simple_linear_regression(input_feature, output):
    n = len(input_feature)
    x = input_feature
    y = output
    #mean of input feature and output 
    x_mean = x.mean()
    y_mean = y.mean()  
    sum_xy = (x*y).sum()
    xy_by_n = (y.sum()*x.sum())/n
    x_square = (x**2).sum()
    xx_by_n = (x.sum()*x.sum())/n

    slope = (sum_xy - xy_by_n)/(x_square-xx_by_n)
    intercept = y_mean - (slope*x_mean)

    return(intercept, slope)
```

```python
input_feature = train_data["sqft_living"]
output = train_data["price"]
```

```python
sqft_intercept, sqft_slope = simple_linear_regression(input_feature, output)
print(sqft_intercept, sqft_slope)
```

```python
def get_regression_predictions(input_feature, intercept, slope):
    predicted_output = intercept + (slope*input_feature)
    return(predicted_output)
```

```python
print(get_regression_predictions(2650, sqft_intercept, sqft_slope))
```

```python
def get_residual_sum_of_squares(input_feature, output, intercept,slope):
    predicted_value = intercept + (slope*input_feature)
    residual = output - predicted_value
    RSS = (residual * residual).sum()
    return(RSS)
```

```python
rss_price = get_residual_sum_of_squares(train_data["sqft_living"], train_data["price"], sqft_intercept, sqft_slope)
print(rss_price)
```

```python
def inverse_regression_predictions(output, intercept, slope):
    estimated_input = (output-intercept)/slope
    return(estimated_input)
```

```python
print(inverse_regression_predictions(800000,sqft_intercept, sqft_slope))
```
